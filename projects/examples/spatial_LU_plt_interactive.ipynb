```python
%matplotlib inline
import warnings
warnings.simplefilter("ignore") # Silence warnings
import hvplot.xarray
import xarray as xr
import cartopy.crs as ccrs

import geoviews as gv
import holoviews as hv

import hvplot.pandas  # noqa
import hvplot.dask  # noq

import panel as pn
import bokeh
from bokeh.resources import INLINE
```

```python
def lu_geo(year):
    ds = xr.open_dataset(f"naturalveg{year}.nc", decode_times=False).load()
    ds['landuse'] = 1.0 - ds['coverfrac']    
    return (ds.isel(time=0).hvplot(x='longitude', y='latitude', z='landuse', cmap = 'Spectral_r', title = f"Land use fraction in {str(year)}", clabel="             Land use fraction"))
```

```python
lu_geo(1900)
```

```python
years      = [1900, 1920, 1940, 1960, 1980, 2000, 2014]
curve_dict = {(y): lu_geo(y) for y in years}
hmap = hv.HoloMap(curve_dict, kdims=['years']).opts(height=400,width=800)

#panel_object = pn.pane.HoloViews(hmap, widget_location='top')
pn.pane.HoloViews(hmap, widget_location='top').save('LU', embed=True)
```
